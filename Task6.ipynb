# Q1. Top ten movies in 2021 according to IMDB Score.

```python
# Filter the DataFrame for movies released in 2021
movies_2021 = df[df['year'] == 2021]

# Sort the movies by score in descending order and select the top 10
top_10_movies_2021 = movies_2021.sort_values(by='imdb_score', ascending=False).head(10)

# Plotting the vertical bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10_movies_2021['title'], top_10_movies_2021['imdb_score'], color='skyblue')
plt.xlabel('Movie Title')
plt.ylabel('Score')
plt.title('Top 10 Movies of 2021 by Score')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Ans. With this Bar chart of Movies release in 2021, we can see that "Seaspiracy", "The Disciple", "The Dig", "Pieces of a Women", "Ferry", "Pele", "Operation Varsity Blues","Pagglait" and "Biggie" are among the top 10 movies and "Seaspiracy" Movie got the highest Score.


#Q2. Top ten English movies in 2021 according to IMDB Score.

```python
# Filter the DataFrame for English movies released in 2021
english_movies_2021 = df[(df['language'] == 'English') & (df['year'] == 2021)]

# Sort the movies by score in descending order and select the top 10
top_10_english_movies = english_movies_2021.sort_values(by='imdb_score', ascending=False).head(10)

# Plotting the vertical bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10_english_movies['title'], top_10_english_movies['imdb_score'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Movie Title')
plt.ylabel('Score')
plt.title('Top 10 English Movies in 2021 by Score')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

# Ans. So in year 2021, "Seaspiracy", "Pieces of a Woman", "The Dig", "The White Tiger", "Operation Varsity Blues", "Pele", "Biggie", "Arlo the Alligator Boy", "Malcom & Marie" and "Moxie" are the top 10 english movies. **Among them "Seaspiracy" Movie score highest in 2021.**

# Q3. Which genres were most popular in 2018th year based on the number of top 10 movies released.

```python
# Filter the DataFrame for movies released in 2018
movies_2018 = df[df['year'] == 2018]

# Sort the movies by score in descending order and select the top 10
top_10_movies_2018 = movies_2018.sort_values(by='imdb_score', ascending=False).head(10)

# Calculate the distribution of movies by genre
genre_counts = top_10_movies_2018['genre'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Top 10 Movies by Genre in 2018')
plt.show()
```

# Ans. With this pie graph we can eaisly understand that "Documentry" movies where more popular in year 2018. This is also reflect audience preferences if the movie industry responds to demand.

# Q4. Show the number of top 10 movies released in each language in year 2020.

```python
# Sort the DataFrame by score in descending order and select the top 10 movies
top_10_movies = df.sort_values(by='imdb_score', ascending=False).head(10)

# Calculate the number of top 10 movies released in each language
language_counts = top_10_movies['language'].value_counts()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(language_counts.index, language_counts.values, color='skyblue')
plt.xlabel('Language')
plt.ylabel('Number of Movies')
plt.title('Number of Top 10 Movies Released in Each Language')
plt.show()
```

# Ans. With this Bar chart we can understand that mostly English movies were released and also show the trends of the movies.

# Q5. In Which Year heightest number of movies released from 2014 to 2021.

```python
# Counting the number of movies released each year
movies_per_year = df['year'].value_counts().sort_index()

# Plotting the yearly trends
plt.figure(figsize=(10, 6))
plt.plot(movies_per_year.index, movies_per_year.values, marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Yearly Trends of Movies Released (2014-2021)')
plt.grid(True)
plt.show()
```

# Ans. With this grid-chart we can easily understand that in year 2020 More numbers of Movies relased, which is more than 175 movies in a year.

# Q6. show how movie lengths vary and highlighting common runtime ranges.

```python
# Plotting the histogram for movie runtimes
plt.figure(figsize=(10, 6))
plt.hist(df['runtime'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Number of Movies')
plt.title('Distribution of Movie Runtimes')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#Ans. This histogram appears bell-shaped; it suggests that most movies have runtimes around a central value, with fewer movies having significantly shorter or longer runtimes. This distribution of runtimes can provide insights into industry trends and audience preferences. For instance, a concentration of movies around certain lengths might suggest optimal durations for audience engagement.
