# **Q1.** How many people have heart disease and how many people doesn't have heart disease?
#Solution:-

```python
#Getting the values:-
df.target.value_counts()
```

```python
#Plotting bar chart:-
df.target.value_counts().plot(kind= 'bar', color=["orchid","salmon"])
plt.title("Heart Disease values")
plt.xlabel("1 = Heart Disease, 0 = No Heart Disease")
plt.ylabel("Amount")
```

```python
#Plotting a pie chart:-
df.target.value_counts().plot(kind='pie', figsize=(8,6))
plt.legend(["1=Disease", "0=No Disease"])
```

#'0' represent 'Female'
#'1' represent 'Male'
# '0' represent 'No Disease'
#'1' represent 'Disease'

#Now let's check how many `Male` and `Female` are in the dataset

```python
df.sex.value_counts()
```

```python
#Plotting a pie chart:-
df.sex.value_counts().plot(kind='pie', figsize=(8,6))
plt.title("Male Female Ratio")
plt.legend(["Mae","Female"])
```

#Let's find the answer of our 2nd Question:-
# **Q2.** People of which sex has most heart disease?

```python
pd.crosstab(df.target, df.sex)
```

```python
sns.countplot(x = 'target', data = df, hue = 'sex')
plt.title("Heart Disease Frequency for sex")
plt.xlabel("0=No Heart Disease, 1=Heart Disease")
```

#Number of male is more than double in our dataset than female.
#More than 45% has disease and 75% female has heart disease.

#Let's move to question 3
# **Q3.** People of Which sex has which type of chest pain most?

```python
#Counting values for different chest pain
df.cp.value_counts()
```

```python
#Plotting a bar chart
df.cp.value_counts().plot(kind='bar', color=['salmon', 'lightskyblue','springgreen','khaki'])
plt.title("Chest pain type vs count")
```

```python
pd.crosstab(df.sex, df.cp)
```

```python
pd.crosstab(df.sex, df.cp).plot(kind='bar', color =['coral', 'lightskyblue', 'plum','khaki'])
plt.title("Type of chest pain for sex")
plt.xlabel('0 = Femal, 1 = Male')
```

#Most of 'Male' has "Type 0" chest pain and least of "Male" has "Type 4" pain.
#In case of "Female" "Type 0" and "Type 2" is percentage is almost same.

#Now let's look at our age column.

```python
#Create a distribution plot with normal distribution curve
sns.displot(x='age', data = df, bins = 30, kde = True)
```

#'58-59' year old people are most in the dataset

```python
#Let's plot another distribution plot for 'Maximum heart rate'
sns.displot(x = 'thalach', data=df, bins=30, kde=True, color='chocolate')
```

#From this plot we get a clear overview about Maximum heart rate represented by "Thalach"

```python
# Visualize cholesterol levels by sex
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='chol', data=df, hue='sex')
plt.title('Cholesterol Levels by Sex')
plt.xlabel('0 = Femal, 1 = Male')
plt.ylabel('Cholesterol Level')
plt.show()
```

#With this box chart we come to know that maximum number of Females suffering from high level of cholesterol.

# Q6. How much variation there is in cholesterol levels.

```python
# Histogram with KDE for cholesterol levels
plt.figure(figsize=(10, 6))
sns.histplot(df['chol'], kde=True, bins=30)
plt.title('Cholesterol Level Distribution')
plt.xlabel('Cholesterol Level')
plt.ylabel('Frequency')
plt.show()
```

#A wide spread suggests diverse cholesterol levels among individuals. Here from graph we can see Large number of people having high level of cholestrol, which is not good for their future.


# Q.7 Which age people mostly suffered with highest heart rate?

```python
sns.jointplot(x='age', y='thalach', data=df, kind='scatter', hue='cp')
plt.suptitle('Joint Plot of Age and Maximum Heart Rate Achieved', y=1.02)
plt.show()
```

# Ans. With this graph we come to know that people with age 35 to 70 suffered high heart rate and this will lead to heart failure.
